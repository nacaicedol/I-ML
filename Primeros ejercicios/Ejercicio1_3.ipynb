a. Si w(t) clasifica mal a x(t), entonces w^T(t)⋅x(t)y(t) tienen diferente signo porque son menores a 0.


b. Demostraciòn:

    x(t)w^T(t+1)y(t) = y(t)(w(t)+y(t)x(t))^T x(t)
    =  y(t)(w^T(t)+y(t)x^T(t)) x(t)
    = x(t)w^T(t)y(t) + y^2(t)x^T(t)x(t)
    como y^2(t)x^T(t)x(t) > 0 entonces
    x(t)w^T(t+1)y(t) > x(t)w^T(t)y(t)

c. Argumentaciòn:

Partimos de que cada paso del algoritmo x(t)w^T(t)y(t) crece linalmente como minimo, y en los casos donde el x(t) esta mal cladificado por w(t), suponemos el primer caso donde y(t) es negativo, pero w^T(t)x(t) es positivo, x(t)w^T(t)y(t) debe incrementar, por lo que w^T(t)x(t) disminuye y se va hacia la region negativa, por lo que es correcto; en el segundo caso tomamos y(t) como positivo, y w^T(t)x(t) como negativo, su producto debe aumentar, por lo que debe moverse hacia la region positiva, por lo que estarìa bien clasificado.
