a.	Palabras que podrían hacer que un correo se considere spam:

i.	Palabras que suponen personalización excesiva

	Felicitaciones
	Has ganado
	Para ti
	Has sido seleccionado
	Querido amigo

ii.	Palabras que inducen a la compra directa

	Aceptamos tarjetas de crédito
	Haz clic en este enlace
	Compra ahora mismo
	Haz clic aquí para comprar

iii.	Palabras que buscan crear una sensación de urgencia

	Hazlo ahora
	Urgente
	Por tiempo limitado
	¿Qué esperas?

iv.	Palabras que generan confianza en el usuario

	100% satisfecho
	Sin riesgos
	Únete a millones de personas
	Satisfacción garantizada
	Esto no es spam


b.	Palabras que generarían peso negativo en el perceptrón

    i. Quien envía el correo, si es una institución acreditada como una universidad o una empresa.

    ii. Cualquier palabra que no sea catalogada como spam, puede tener un peso negativo u cero.


c.	Los pesos w, que marquen las palabras asociadas con el spam, junto con el par ordenado de datos de entrenamiento y salida.
